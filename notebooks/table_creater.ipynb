```python
from pathlib import Path

import pandas as pd
```

```python
results = pd.DataFrame({
    "Method": ["Baseline", "VAE", "ICA", "Ours"],
    "Log-Likelihood": [-1.234, -1.102, -1.087, -0.943],
    "MCC": [0.412, 0.587, 0.621, 0.794],
})
results
```

```python
def df_to_latex_body(df: pd.DataFrame, float_fmt: str = "{:.3f}") -> str:
    r"""Render only the row body (no tabular env, no rules) for \\input.

    The final row is emitted *without* a trailing ``\\\\``: that closing
    row-break belongs in the parent .tex file, immediately before
    ``\\bottomrule``. Putting it inside the included file leaves a stray
    token between ``\\\\`` and ``\\noalign`` that breaks booktabs rules.
    """
    formatted = []
    for _, row in df.iterrows():
        formatted.append(" & ".join(float_fmt.format(v) if isinstance(v, float) else str(v) for v in row))
    return " \\\\\n".join(formatted) + "\n"


body = df_to_latex_body(results)
print(body)
```

```python
out_path = Path("./paper/tables/results.tex")
out_path.parent.mkdir(parents=True, exist_ok=True)
out_path.write_text(data=body, encoding="utf-8")
print(f"wrote {out_path} ({len(body)} bytes)")
```
